```python
def supermarket(x):
    i=0
    while i<240:
        if str(x) == df1.at[i,'location']:
            print(df1.at[i,'Supermarket'])
        i=i+1
    return
```

```python
supermarket('Yelahanka')
```

```python
supermarket('Yeshwanthpur')
```
